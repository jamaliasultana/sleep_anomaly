```python
# Filter the first 10 participants based on 'pid' for all datasets
pids = [f'INS-W_00{i}' for i in range(1, 11)]

# Filter sleep_csv for the first 10 participants
sleep_csv_10_participants = sleep_csv[sleep_csv['pid'].isin(pids)]

# Filter steps_csv and screen_csv for the first 10 participants
steps_csv_10_participants = steps_csv[steps_csv['pid'].isin(pids)]
screen_csv_10_participants = screen_csv[screen_csv['pid'].isin(pids)]

# Merge the datasets based on 'pid'
merged_df = pd.merge(sleep_csv_10_participants, steps_csv_10_participants, on='pid')
merged_df = pd.merge(merged_df, screen_csv_10_participants, on='pid')

# Display the merged data
print(merged_df)

# Optionally, you can save the merged data to a CSV file
merged_df.to_csv('merged_10_participants_steps_sleeps.csv', index=False)
print("Merged data for the first 10 participants has been saved as merged_10_participants.csv")
```
